```python
import math
import pandas as pd
import seaborn as sb
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import norm
```

```python
def mean(data):
    n = len(data)
    total = sum(data)
    return total/n

def sd(data):
    n = len(data)
    
    if n <=1:
        return 0.0
    
    m = mean(data)
    sd = 0.0
    
    for d in data:
        sd += (float(d) - m)**2
        
    sd = math.sqrt(sd/float(n-1))
    return sd

r = norm.rvs(size=1000)

print(mean(r))
print(sd(r))

print(np.mean(r))
print(np.std(r))

plt.hist(r)
```

```python
path = "/Users/derekolson/DerekOlson/CS6017/homework1/2019-PM2.5.csv"
data = pd.read_csv(path)

series = data[['Unnamed: 0', 'SM.2']]
series.columns = ['Date', 'SM.2']
series = series.dropna()

df = series.astype({'Date':'datetime64[ns]', 'SM.2':'float64'})
df.index = df['Date']

plt.plot(df['Date'], df['SM.2'])
plt.show()
```

A lot of noise

```python
monthly = df.resample('M').mean()
monthly['Date'] = monthly.index
monthly['Month'] = monthly['Date'].dt.month
sb.barplot(x='Month', y='SM.2', data=monthly)
```

Difficult to understand the within month distribution. Potentially greater variability in Jan and Dec.

```python
hourly = df.groupby(df.index.hour).mean()
hourly['Hour'] = hourly.index
sb.barplot(x='Hour', y='SM.2', data=hourly)
```

Seems to be less pollution in the afternoon.

```python
df['month'] = df['Date'].dt.month
sb.boxplot(x='month', y='SM.2', data=df)
```

December and January seem substantially differnet than other months. Medians are outside of upper quartile of other months. 

```python
df['hour'] = df['Date'].dt.hour
sb.boxplot(x='hour', y='SM.2', data=df)
```

Difficult to vizualize significant diurnal trends, but seems to be some decline in particulates in afternoon. Median values are quite similar across all hours.
